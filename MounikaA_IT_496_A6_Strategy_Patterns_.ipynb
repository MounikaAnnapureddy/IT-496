```python
import pandas as pd
import math
# create a dataframe for our alternative analysis matrix
aam =pd.DataFrame(columns=['criteria','weight','ratingA','ratingB','scoreA','scoreB'])
print(aam)
```

```python
#first step: add a list of criteria
aam['criteria']=['Cost Savings','Profitability','Risk to organization','Risk to Customers']
aam['weight']=[0.4,0.2,0.2,0.3]
aam['ratingA']=[10,7,6,8]
aam['ratingB']=[10,2,8,10]
#step 4: calculate partial scores by multiplying weight * ratings
for index,row in aam.iterrows():
  aam['scoreA'][index]=row['ratingA']*row['weight']
  aam['scoreB'][index]=row['ratingB']*row['weight']

print(aam)
```

```python
totalScoreA=0
totalScoreB=0
for index, row in aam.iterrows():
  totalScoreA+=row['scoreA']
  totalScoreB+=row['scoreB']

print ('the total score for A 1s {:.2f} and for B is {:.2f}'.format (totalScoreA, totalScoreB))
```

```python
import numpy as np
import matplotlib.pyplot as plt
axisLocations = np.linspace(start=0, stop=2 * np.pi, num=len(aam['ratingA']), endpoint=False)
axisLocations = np.concatenate((axisLocations,[axisLocations[0]]))
ratingsA=aam['ratingA']
ratingsB=aam['ratingB']
ratingsACircular=np.concatenate((ratingsA,[ratingsA[0]]))
ratingsBCircular=np.concatenate((ratingsB,[ratingsB[0]]))
plt.figure(figsize=(8,8))
plt.subplot(polar=True)
plt.plot(axisLocations,ratingsACircular,'r-', linewidth=2, label='In-house development')
plt.plot(axisLocations,ratingsBCircular,'b-', linewidth=2, label='outsource')
plt.title('Project options comparison', size=40)
lines, labels =plt.thetagrids(np.degrees(axisLocations), labels=aam['criteria'])
plt.legend()
plt.show()
```
