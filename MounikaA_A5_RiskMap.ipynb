```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
riskDF=pd.DataFrame(columns=['risk','likelihood','impact','score'])
#'loss of systems control', 'Distributed Denial-of-Service (DDoS) attack', 'access control and fraud detection', 'data breach by means of internet connectivity'
riskDF['risk']=['loss of systems control', 'access control and fraud detection', 'DDoS attack', 'data breach by means of internet connectivity']
riskDF['likelihood']=[3.6,1.8,0.75,4.5]
riskDF['impact']=[8,3,6,9]
riskDF['score']=riskDF['impact']*riskDF['likelihood']
scoresPivot=riskDF.pivot('impact','likelihood','score')
print(scoresPivot)
labelsPivot=riskDF.pivot('impact','likelihood','risk')
print(labelsPivot)
sns.set(rc={'figure.figsize':(16,6)})
p1=sns.heatmap(scoresPivot, cmap='Pastel2_r', annot=labelsPivot, fmt='')
```
